# How to create an AWS Role using OneCloud AWS Access Management:
1. Go to **OneCloud AWS Access Management** and select the proper **AWS account** found in the left-hand panel under **My provisioned accounts**.
2. Make sure you're on the **Accounts** tab and scroll down and select **Roles**.





3. Select **Create custom role**.



4. Enter a **Role Name** and select what sort of **Purpose** is appropriate. If creating a role for an AWS service (like Lambda), select **Service**.





5. Search for the appropriate **Service Namespace** if a **Service** **Purpose** is selected.



6. Press **Create Role**.
7. Exit **OneCloud** and go to the associated **AWS Console** and go to the **IAM** module.
8. Select **Roles** on the left panel and search for associated role name from step 4.



9. Under **Permissions policies**, select **Add permissions** and add the appropriate permissions based off its use case.
