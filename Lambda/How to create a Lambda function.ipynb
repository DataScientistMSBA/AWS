# How to create a Lambda function:
1. In AWS, go to the Lambda module anc click on **Create Function**.
2. Select **Author from scratch**, give it a **Function name** and select a Python based **Runtime**.



3. Open up the **Change default execution role** and select **Use an existing role**. Note, if you have no existing roles to select from then you will need to create one. Instruction on this can be found in my AWS repo under the following: [IAM/Roles/How to create an AWS Role using OneCloud AWS Access Management.ipynb]()



4. Select the appropriate **Existing role** and click **Create function**.
